```python
import json
import urllib.request as url
```

```python
res = url.urlopen("https://api.covid19india.org/states_daily.json")
```

```python
res
```

```python
data = json.load(res)
```

```python
type(data)
```

```python
data.keys()
```

```python
data = data['states_daily']
```

```python
type(data)
```

```python
# data[0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df.shape
```

```python
df.columns
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
confirmed_df = df[df['status'] == 'Confirmed']
```

```python
import matplotlib.pyplot as plt
```

```python
df['dl'][0]
```

```python
df['dl'] = df['dl'].astype('int32')
```

```python
df['dl'][0]
```

```python
plt.scatter(confirmed_df['date'], confirmed_df['dl'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(confirmed_df['date'], confirmed_df['dl'])
plt.xticks(rotation='vertical')
plt.title("Confirmed cases in delhi")
plt.show()
```

```python
df['up'] = df['up'].astype('int32')
confirmed_df = df[df['status'] == 'Confirmed']
```

```python
plt.figure(figsize=(10,6))
plt.plot(confirmed_df['date'], confirmed_df['up'])
plt.xticks(rotation='vertical')
plt.title("Confirmed cases in UP")
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(confirmed_df['date'], confirmed_df['dl'],label='Delhi')
plt.plot(confirmed_df['date'], confirmed_df['up'],label='UP')
plt.xticks(rotation='vertical')
plt.title("Confirmed cases in Delhi vs UP")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(confirmed_df['date'], confirmed_df['up'])
plt.scatter(confirmed_df['date'], confirmed_df['dl'])
plt.xticks(rotation='vertical')
plt.title("Confirmed cases in UP")
plt.show()
```
